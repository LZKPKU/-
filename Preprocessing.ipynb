```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
% matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
data = pd.read_csv("1_0.csv")
print(data.shape)
data.head(5)
```

```python
data = data.drop(data[(data['GrLivArea']>4000) & (data['SalePrice']<300000)].index)
print(data.shape)
```

```python
data['BsmtQual'].fillna('NoBasement', inplace=True)
data['bsmt_height'] = 0
data.loc[(data.BsmtQual == 'Fa'), 'bsmt_height'] = 1
data.loc[(data.BsmtQual == 'TA'), 'bsmt_height'] = 2
data.loc[(data.BsmtQual == 'Gd'), 'bsmt_height'] = 3
data.loc[(data.BsmtQual == 'Ex'), 'bsmt_height'] = 4
pd.crosstab(data.BsmtQual, data.bsmt_height)
```

```python
data.MasVnrArea.describe()
```

```python
data['exterior_quality'] = 0
data.loc[(data.ExterQual == 'TA'), 'exterior_quality'] = 1
data.loc[(data.ExterQual == 'Gd'), 'exterior_quality'] = 2
data.loc[(data.ExterQual == 'Ex'), 'exterior_quality'] = 3
pd.crosstab(data.ExterQual, data.exterior_quality)
```

```python
data['exterior_condition'] = 0
data.loc[(data.ExterCond == 'Fa'), 'exterior_condition'] = 1
data.loc[(data.ExterCond == 'TA'), 'exterior_condition'] = 2
data.loc[(data.ExterCond == 'Gd'), 'exterior_condition'] = 3
data.loc[(data.ExterCond == 'Ex'), 'exterior_condition'] = 4
pd.crosstab(data.ExterCond, data.exterior_condition)
```

```python
data[['exterior_quality','exterior_condition']].corr()  # Correlation is just 0.01, good!
```

```python
N = 1458
data['BsmtCond'].fillna('NoBasement', inplace=True)
print((data.BsmtCond.value_counts().sort_values(ascending = False)/N*100).round(2))
```

```python
data['bsmt_condition'] = 0
data.loc[(data.BsmtCond == 'Po'), 'bsmt_condition'] = 1
data.loc[(data.BsmtCond == 'Fa'), 'bsmt_condition'] = 2
data.loc[(data.BsmtCond == 'TA'), 'bsmt_condition'] = 3
data.loc[(data.BsmtCond == 'Gd'), 'bsmt_condition'] = 4
pd.crosstab(data.BsmtCond, data.bsmt_condition)
```

```python
data['BsmtExposure'].fillna('NoBasement', inplace=True)
print((data.BsmtExposure.value_counts().sort_values(ascending = False)/N*100).round(2))
```

```python
data['bsmt_exposure'] = 0
data.loc[(data.BsmtExposure == 'Mn'), 'bsmt_exposure'] = 1
data.loc[(data.BsmtExposure == 'Av'), 'bsmt_exposure'] = 2
data.loc[(data.BsmtExposure == 'Gd'), 'bsmt_exposure'] = 3
pd.crosstab(data.BsmtExposure, data.bsmt_exposure)
```

```python
data['BsmtFinType1'].fillna('NoBasement', inplace=True)
data['BsmtFinType2'].fillna('NoBasement', inplace=True)
data['bsmt_finished1'] = 0
data.loc[(data.BsmtFinType1 == 'Unf'), 'bsmt_finished1'] = 1
data.loc[(data.BsmtFinType1 == 'LwQ'), 'bsmt_finished1'] = 1
data.loc[(data.BsmtFinType1 == 'BLQ'), 'bsmt_finished1'] = 2
data.loc[(data.BsmtFinType1 == 'Rec'), 'bsmt_finished1'] = 2
data.loc[(data.BsmtFinType1 == 'ALQ'), 'bsmt_finished1'] = 3
data.loc[(data.BsmtFinType1 == 'GLQ'), 'bsmt_finished1'] = 4
pd.crosstab(data.BsmtFinType1, data.bsmt_finished1)
```

```python
data['bsmt_finished2'] = 0
data.loc[(data.BsmtFinType2 == 'Unf'), 'bsmt_finished2'] = 1
data.loc[(data.BsmtFinType2 == 'LwQ'), 'bsmt_finished2'] = 1
data.loc[(data.BsmtFinType2 == 'BLQ'), 'bsmt_finished2'] = 2
data.loc[(data.BsmtFinType2 == 'Rec'), 'bsmt_finished2'] = 2
data.loc[(data.BsmtFinType2 == 'ALQ'), 'bsmt_finished2'] = 3
data.loc[(data.BsmtFinType2 == 'GLQ'), 'bsmt_finished2'] = 4
pd.crosstab(data.BsmtFinType2, data.bsmt_finished2)
```

```python
data['BsmtFinishedShare'] = 1 - data['BsmtUnfSF']/data['TotalBsmtSF']
data.loc[data.TotalBsmtSF == 0, 'BsmtFinishedShare'] = 0
```

```python
data['heating_quality'] = 0
data.loc[data.HeatingQC == 'Fa', 'heating_quality'] = 1
data.loc[data.HeatingQC == 'TA', 'heating_quality'] = 2
data.loc[data.HeatingQC == 'Gd', 'heating_quality'] = 3
data.loc[data.HeatingQC == 'Ex', 'heating_quality'] = 4
pd.crosstab(data.HeatingQC, data.heating_quality)
```

```python
data['kitchen_quality'] = 0
data.loc[data.KitchenQual == 'TA', 'kitchen_quality'] = 1
data.loc[data.KitchenQual == 'Gd', 'kitchen_quality'] = 2
data.loc[data.KitchenQual == 'Ex', 'kitchen_quality'] = 3
pd.crosstab(data.KitchenQual, data.kitchen_quality)
```

```python
data['home_funct'] = 0
data.loc[data.Functional == 'Sev', 'home_funct'] = 0
data.loc[data.Functional == 'Maj2', 'home_funct'] = 1
data.loc[data.Functional == 'Maj1', 'home_funct'] = 1.5
data.loc[data.Functional == 'Mod', 'home_funct'] = 3
data.loc[data.Functional == 'Min2', 'home_funct'] = 5
data.loc[data.Functional == 'Min1', 'home_funct'] = 5.5
data.loc[data.Functional == 'Typ', 'home_funct'] = 7
pd.crosstab(data.Functional, data.home_funct)
```

```python
# Creating a quantitative variable - treating "no fireplace" as the worst level (zero)
data['fireplace_quality'] = 0
data.loc[data.FireplaceQu == 'Po', 'fireplace_quality'] = 1
data.loc[data.FireplaceQu == 'Fa', 'fireplace_quality'] = 1  
data.loc[data.FireplaceQu == 'TA', 'fireplace_quality'] = 2
data.loc[data.FireplaceQu == 'Gd', 'fireplace_quality'] = 3
data.loc[data.FireplaceQu == 'Ex', 'fireplace_quality'] = 3.5
pd.crosstab(data.FireplaceQu, data.fireplace_quality)
```

```python
data.loc[(data.Fireplaces>0)&(data.fireplace_quality == 0), 'fireplace_quality'] = None
```

```python
data['fireplace_ttlscore'] = data['Fireplaces'] * data['fireplace_quality']
```

```python
data['gar_type'] = 0
data.loc[data.GarageType == 'CarPort', 'gar_type'] = 1
data.loc[data.GarageType == 'Detchd', 'gar_type'] = 1.5
data.loc[data.GarageType == 'Attchd', 'gar_type'] = 2
data.loc[data.GarageType == 'Basment', 'gar_type'] = 2   # same as Attached
data.loc[data.GarageType == 'BuiltIn', 'gar_type'] = 2.5
data.loc[data.GarageType == '2Types', 'gar_type'] = 3
pd.crosstab(data.GarageType, data.gar_type)
```

```python
# Creating a quantitative variable - treating "no garage" as the worst level (zero)
data['gar_finish'] = 0
data.loc[data.GarageFinish == 'Unf', 'gar_finish'] = 1
data.loc[data.GarageFinish == 'RFn', 'gar_finish'] = 2
data.loc[data.GarageFinish == 'Fin', 'gar_finish'] = 3
pd.crosstab(data.GarageFinish, data.gar_finish)
```

```python
data['gar_qual'] = 3
data.loc[data.GarageQual == 'Po', 'gar_qual'] = 1
data.loc[data.GarageQual == 'Fa', 'gar_qual'] = 2
data.loc[data.GarageQual == 'Gd', 'gar_qual'] = 3.5
data.loc[data.GarageQual == 'Ex', 'gar_qual'] = 4
pd.crosstab(data.GarageQual, data.gar_qual)
```

```python
data['gar_cond'] = 2
data.loc[data.GarageCond == 'Po', 'gar_cond'] = 1
data.loc[data.GarageCond == 'Fa', 'gar_cond'] = 1.5
data.loc[data.GarageCond == 'Gd', 'gar_cond'] = 2.5
data.loc[data.GarageCond == 'Ex', 'gar_cond'] = 3
pd.crosstab(data.GarageCond, data.gar_cond)
```

```python
data['gar_perception'] = data['gar_qual'] * data['gar_cond']
```

```python
data['driveway'] = 0
data.loc[data.PavedDrive == 'P', 'driveway'] = 2
data.loc[data.PavedDrive == 'Y', 'driveway'] = 3
pd.crosstab(data.PavedDrive, data.driveway)
```

```python
data['fence'] = 0  # No fence are all zero 
data.loc[(data.Fence == 'MnWw')|(data.Fence == 'MnPrv'), 'fence'] = 0.5 # Minimum Wood or Minimum Privacy are 0.5
data.loc[(data.Fence == 'GdWo')|(data.Fence == 'GdPrv'), 'fence'] = 1.5
pd.crosstab(data.Fence, data.fence)
```

```python
data['full_baths'] = data['BsmtFullBath'] + data['FullBath']
data['half_baths'] = data['BsmtHalfBath'] + data['HalfBath']
```

```python
temp = data[['exterior_quality', 'heating_quality', 'kitchen_quality', 'OverallQual', 'bsmt_height']]
scaler = StandardScaler()
temp_scaled = pd.DataFrame(scaler.fit_transform(temp), columns=temp.columns)
data['average_quality'] = temp_scaled.mean(axis = 1)
```

```python
data['room_size'] = data['GrLivArea'] / data['TotRmsAbvGrd']
```

```python
data['age'] = data['YrSold'] - data['YearBuilt']
```

```python
data['garage_perception'] = (data['gar_type'] + 2*data['GarageCars'] + data['gar_finish'])/3
```

```python
quant_vars_all = data.select_dtypes(include=['number'])
cat_vars_all = data.select_dtypes(exclude=['number'])
print(quant_vars_all.shape)
print(cat_vars_all.shape)
```

```python
data['LogSalePrice'] = np.log(data['SalePrice'])
```

```python
data = data.drop(['Utilities', 'RoofMatl','Heating','MiscFeature'],axis=1)
```

```python
data = data.drop(['BsmtQual','ExterQual','ExterCond', 'BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','HeatingQC',
'KitchenQual','Functional','FireplaceQu','Fireplaces','GarageType','GarageFinish','GarageQual','GarageCond',
'PavedDrive','Fence','BsmtFullBath','BsmtHalfBath', 'YearBuilt','gar_type','GarageCars','gar_finish'], axis=1)
```
